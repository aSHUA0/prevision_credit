**Histogramas**

```python
import os
import pandas as pd
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

caminho = os.path.join("..", "data", 'UCC.csv')

credit = pd.read_csv(caminho)

#credit = Banco de dados usado
```

```python
sns.histplot(credit['AGE'], bins=30, kde=False, color = 'blue')
plt.title('Histograma')
plt.xlabel('Idade')
plt.ylabel('Quantidade')

plt.show()
```

```python
sns.histplot(credit['EDUCATION'], bins=30, kde=False, color = 'blue')
plt.title('Histograma')
plt.xlabel('Educa√ß√£o')
plt.ylabel('Quantidade')

plt.show()
```

```python
sns.histplot(credit['SEX'], bins=15, kde=False, color = 'blue')
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.title('Histograma')
plt.xlabel('Gen√™ro')
plt.ylabel('Quantidade')

plt.show()
```

```python
sns.histplot(credit['MARRIAGE'], bins=30, kde=False, color = 'blue')
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.title('Histograma')
plt.xlabel('Relacionamento')
plt.ylabel('Quantidade')

plt.show()
```

```python
fig, ax = plt.subplots()

PAY = credit[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']]
ax.boxplot([PAY[col] for col in PAY.columns], positions=[2, 4, 6, 8, 10, 12], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})


plt.show()
```

```python
fig, ax = plt.subplots()

PAY_AMT = credit[['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']]
ax.boxplot([PAY_AMT[col] for col in PAY_AMT.columns], positions=[2, 4, 6, 8, 10, 12], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

plt.xticks([2, 4, 6, 8, 10, 12])
plt.show()
```
